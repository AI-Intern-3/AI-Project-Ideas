```python
# Creating a list
my_list = [1, 2, 3, 4, 5]
print("Original List:")
print(my_list)

# 1. Appending Items
my_list.append(6)
print("\nAfter appending 6:")
print(my_list)

# 2. Inserting Items
my_list.insert(2, 99)
print("\nAfter inserting 99 at index 2:")
print(my_list)

# 3. Extending a List
my_list.extend([7, 8, 9])
print("\nAfter extending with [7, 8, 9]:")
print(my_list)

# 4. Removing Items
my_list.remove(99)
print("\nAfter removing 99:")
print(my_list)

# 5. Popping Items
popped_item = my_list.pop(3)
print(f"\nAfter popping item at index 3 (popped item: {popped_item}):")
print(my_list)

# 6. Indexing
index_of_six = my_list.index(6)
print(f"\nIndex of item 4: {index_of_six}")

# 7. Slicing
sliced_list = my_list[5:7]
print("\nSlicing from index 1 to 3:")
print(sliced_list)

# 8. Sorting
sorted_list = sorted(my_list)
print("\nSorted List:")
print(sorted_list)

# Sorting the list in place
my_list.sort(reverse=True)
print("\nList sorted in place in descending order:")
print(my_list)

# 9. Reversing
my_list.reverse()
print("\nList reversed:")
print(my_list)

# 10. Clearing
my_list.clear()
print("\nList cleared:")
print(my_list)

# Re-populate list for further operations
my_list = [1, 2, 3, 4, 5]

# 11. Length of the list
length = len(my_list)
print("\nLength of the list:")
print(length)

# 12. Check if an item exists in the list
exists = 3 in my_list
print("\nDoes 3 exist in the list?")
print(exists)

# 13. Iterating through a list
print("\nIterating through the list:")
for item in my_list:
    print(item, end=" ")

# 14. List comprehensions
squared_list = [x**2 for x in my_list]
print("\n\nList comprehensions (squared values):")
print(squared_list)

# 15. Copying a list
copied_list = my_list.copy()
print("\nCopied List:")
print(copied_list)
```
