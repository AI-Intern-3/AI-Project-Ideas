```python
# Creating a dictionary
my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
print("Original Dictionary:")
print(my_dict)

# 1. Accessing Values
print("\nAccessing Values:")
print("Name:", my_dict['name'])
print("Age:", my_dict.get('age'))

# 2. Adding or Updating Items
my_dict['email'] = 'john@example.com'  # Adding
my_dict['age'] = 31  # Updating
print("\nAfter Adding/Updating Items:")
print(my_dict)

# 3. Removing Items
removed_item = my_dict.pop('city')  # Removing an item and getting its value
print("\nAfter Removing 'city':")
print(my_dict)
print("Removed Item:", removed_item)

del my_dict['email']  # Another way to remove an item
print("\nAfter Removing 'email':")
print(my_dict)

# 4. Checking if Key Exists
print("\nChecking if Key Exists:")
print("Is 'name' in the dictionary?", 'name' in my_dict)
print("Is 'city' in the dictionary?", 'city' in my_dict)

# 5. Iterating through Keys, Values, and Items
print("\nIterating through Keys, Values, and Items:")
print("Keys:")
for key in my_dict.keys():
    print(key, end=" ")
print("\nValues:")
for value in my_dict.values():
    print(value, end=" ")
print("\nItems:")
for key, value in my_dict.items():
    print(f"{key}: {value}")

# 6. Getting Keys and Values
keys = list(my_dict.keys())
values = list(my_dict.values())
print("\nGetting Keys and Values:")
print("Keys:", keys)
print("Values:", values)

# 7. Merging Dictionaries
other_dict = {'city': 'Los Angeles', 'phone': '123-456-7890'}
my_dict.update(other_dict)
print("\nAfter Merging with another dictionary:")
print(my_dict)

# 8. Clearing the Dictionary
my_dict.clear()
print("\nAfter Clearing the Dictionary:")
print(my_dict)

# 9. Copying the Dictionary
my_dict = {'name': 'John', 'age': 30, 'address': 'Pune'}
copied_dict = my_dict.copy()
print("\nCopied Dictionary:")
print(copied_dict)

# 10. Using get Method
value = my_dict.get('name')
missing_value = my_dict.get('address', 'Not Found')
print("\nUsing get Method:")
print("Value for 'name':", value)
print("Value for 'address':", missing_value)
```
