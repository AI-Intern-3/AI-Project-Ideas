```python
import numpy as np
import matplotlib.pyplot as plt

# Define the entropy function
def entropy(p):
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

# Handle the case where p is 0 or 1 to avoid log(0)
def safe_entropy(p):
    return entropy(p) if 0 < p < 1 else 0

# Generate probability values
p_values = np.linspace(0, 1, 100)
entropy_values = [safe_entropy(p) for p in p_values]

# Assume a split where we have subsets with different probabilities
# Example: parent node with p = 0.5 splits into two child nodes with p1 and p2
p_parent = 0.5
p_child1 = 0.3
p_child2 = 0.7
w1 = 0.6  # weight of child 1 (proportion of instances in child 1)
w2 = 0.4  # weight of child 2 (proportion of instances in child 2)

entropy_parent = safe_entropy(p_parent)
entropy_child1 = safe_entropy(p_child1)
entropy_child2 = safe_entropy(p_child2)

# Calculate weighted average entropy of children
weighted_entropy_children = w1 * entropy_child1 + w2 * entropy_child2

# Calculate information gain
information_gain = entropy_parent - weighted_entropy_children

# Plot the entropy function
plt.figure(figsize=(12, 6))

# Plot entropy
plt.plot(p_values, entropy_values, label='Entropy H(p)')
plt.axvline(x=p_parent, color='r', linestyle='--', label=f'Parent Node Entropy: H({p_parent})')
plt.axvline(x=p_child1, color='g', linestyle='--', label=f'Child 1 Entropy: H({p_child1})')
plt.axvline(x=p_child2, color='b', linestyle='--', label=f'Child 2 Entropy: H({p_child2})')

# Annotate information gain
plt.text(0.5, 0.8, f'Information Gain = {information_gain:.2f}', fontsize=12, color='purple')

plt.title('Entropy and Information Gain')
plt.xlabel('Probability of Class 1 (p)')
plt.ylabel('Entropy H(p)')
plt.legend()
plt.grid(True)
plt.show()
```
